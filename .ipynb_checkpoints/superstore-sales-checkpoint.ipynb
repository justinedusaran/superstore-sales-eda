### Superstore Sales Data Analysis

This project explores sales trends using Python and Pandas.  
The goal is to identify key sales patterns, top-performing products, and seasonal trends.

```python
#libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#dataset
data = pd.read_csv("superstore-sales-dataset.csv")
data.head()
```

```python
#data types
data.info()
```

```python
#description
data.describe()
```

### Data Cleaning

```python
#remove missing values
data_c = data.dropna()

#fill missing values with column mean
data.fillna(data.select_dtypes(include=["number"]).mean(), inplace=True)

#remove duplicates
data.drop_duplicates(inplace=True)
```

```python
#checking if data is cleaned
data.isnull().sum()
```

### Data Visualization

```python
#sales
plt.figure(figsize=(7,4))
sns.histplot(data["Sales"], bins=50, kde=True)
plt.title("Sales Distribution")
plt.show()
```

#### Analysis 

- The histogram shows the sales transaction having majority of sales below 5000. The distribution is right-skewed, meaning most sales are small, but a few very high sales create a long tail on the right. This shows that while most transactions are low, some big sales significantly impact total revenue.

```python
#sales trend over time
data["Order Date"] = pd.to_datetime(data["Order Date"], format="%d%m%Y")

#group by month
data.groupby(data["Order Date"].dt.to_period("M"))["Sales"].sum().plot(kind="line", figsize=(10,5))
plt.title("Sales Trend Over Time")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

#### Analysis

- The line graph shows fluctuations in sales over time, with noticeable peaks and dips. Sales tend to be higher in Q4 (October–December), likely due to seasonal demand or holiday shopping, while Q1 (January–March) experiences lower sales, possibly due to a post-holiday slowdown. Despite these fluctuations, the overall trend indicates a gradual increase in total sales over time.

```python
#best-selling products
data.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)
```

#### Analysis

- Based on the data above, the Canon imageCLASS 2200 Advanced Copier has the highest sales at ₱61,599.82, followed by the Fellowes PB500 Electric Punch Binding Machine at ₱27,453.38. Other top-selling products include the Cisco TelePresence System EX90 and HON 5400 Series Task Chairs, both exceeding ₱21,000. Most best-sellers are office equipment and binding systems, indicating strong demand in corporate and business settings.

```python
#most profitable product categories
data.groupby("Category")["Sales"].sum().sort_values(ascending=False).plot(kind="bar", figsize=(8,5))
plt.title("Total Sales by Category")
plt.ylabel("Total Sales")
plt.show()
```

#### Analysis

- The bar graph shows total profit by category, with technology being the most profitable, exceeding ₱800,000, while office supplies generate the least profit, ranging between ₱600,000 and ₱700,000.

```python
#cities with highest sales
data.groupby("City")["Sales"].sum().sort_values(ascending=False).head(5)
```

#### Analysis

- Based on the data above, New York City has the highest sales at ₱252,462.55, followed by Los Angeles with ₱173,420.18. Seattle, San Francisco, and Philadelphia also show strong sales, each exceeding ₱100,000.
